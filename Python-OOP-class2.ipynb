```python
def log_calls(fn):
    from time import ctime
    def wrapper(*args, **kwargs):
        with open("log_calls.log", "a") as logfile:
            print("{}: invoking {}({}, {})".format(ctime(),
                                                   fn.__qualname__,
                                                   args,
                                                   kwargs), file=logfile)
        ret = fn(*args, **kwargs)
        print("{}: {}({}, {}) returned {}".format(ctime(), fn.__qualname__, args, kwargs, ret))
        return ret
    return wrapper

@log_calls
def square(x): return x*x

@log_calls
def greet():
    print("Hello world")

square(2)
square(3)
greet()
    
    
```

```python
from functools import wraps
profile_log ={}
def profile_function(fn):
    from time import time
    @wraps(fn)                                            ### to retain the metadata
    def wrap(*args, **kwargs):
        start = time()
        ret = fn(*args, **kwargs)
        end = time()
        duration = end - start
        profile_log[fn] = duration
    return wrap

@profile_function
#@log_calls                                           ## @wraps does not work with more than 1 chain of decorator.
def greet():
    print("Hello World")
    
greet()
print(profile_log)
```

```python
def style(p):
    def decorator(fn):
        if (p == "upper"):            ## this checks should be there in decorator and not in wrapper
            def wrapper(*args, **kwargs):
                return fn(*args, **kwargs).upper()
        elif (p == "bold"):
            def wrapper(*args, **kwargs):
                return "<bold>" + fn(*args, **kwargs) + "</bold>"
        return wrapper
    return decorator


@style("upper")
def greet():
    return "Helo world"

@style("bold")
@style("upper")
def welcome():
    return "welcome to python"

print(greet())
print(welcome())
```

```python
class CommandDispatch:
    def __init__(self):
        self.map = {}
        
    def action(self, cmd):
        def decorate(fn):
            self.map[cmd] = fn
        return decorate
    
    def invalid(self, fn):
        self.invalid_command = fn
    
    def input(self, fn):
        self.input_fn = fn
        
    def run(self):
        while True:
            args = self.input_fn()
            if (type(args) is not list):
                raise TypeError("{} must return a list".format(self.input_fn.__qualname__))
            self.map.get(args[0], self.invalid_fn)(* args)


## from command_dispatch import CommandDispatch

command = CommandDispatch()


@command.action("list")
def list_directory(*args):
    from os import listdir
    if len(args) < 2:
        args += ".",
    for path in args[1:]:
        print("{}:".format(path))
        print("\n".join(listdir(path)))


@command.action("whoami")
def show_user(*args):
    from getpass import getuser
    print(getuser())


@command.action("date")
def print_date(*args):
    from time import ctime
    print(ctime())


@command.action("exit")
def exit_shell(*args):
    exit(0)


@command.invalid
def invalid_command(*args):
    print("Invalid command - ", args[0])


@command.input ## not working fix it.
def get_input():
    return input("PyShell> ").split()


#if __name__ == '__main__':
#    command.run()

command.run()

```

```python

#### to remove based on priority.  ### default is the minheap
from heapq import heappush, heappop    ### this is much faster.
a = []

while True:
    v = input("Enter value: ")
    if not v: break
    p = int(input("Enter prio: "))
    heappush(a, (p, v))
while a:
    print("popped", heappop(a))
```

```python
### MULTIPLE INHERITANCE
## deadly diamond problem -> 

### so if we inherit everything from the all the classes -> it becoem design disaster. 
## Java 8 -> also has multiple inheritance. 

## In Python -> we call it as mixels 
# class inheriting from base class and will also add a "mixin" 

# like https is a subclass of http and SSL.
## it inherits http and mixel the SSL. 
## secure connection. are doen by SSL. 
# ESMTP -> which is a derivative of SMTP and it also mixins SSL. 

### for .e.g. rectangular are mixins. 
# mixins are not instantiated. rectancles shape cannot be found. IT is an abstract -> it is a mixin to some complex item which is rectangular in shape.

## so mixins are kind of abstraction. 

```

```python
# ElectronicDevice is a mixin becase all the devices are electronic device but it on itself is nothing but an abstraction
class ElectronicDevice:
    def poweron(self):
        pass
    def poweroff(self):
        pass
    def buy(self):
        print("inside ElectronicDEvice", self.__class__)
        pass
        
        
## Musicplayer class
class MusicPlayer:
    def play(self):
        pass
    def stop(self):
        pass
    
    def buy(self):
        print("inside Musicplayer", self.__class__)
        pass

# baseclass 1, baseclass 2 
## mixins generally shoudl come 1st and then the actual concrete implementation. => because monkey patching can be done a lot.
# but no hard and fast rule. 

class MP3Player(ElectronicDevice, MusicPlayer):
    pass

m = MP3Player()
m.play()
m.stop()
m.buy()       ### -> this will 1st check the base class1 then base class2 ->happens at runtime.


## if we want musicplayer buy to be called then 
# defien buy inside MP3Player  method and call Musicplayer.buy(self) ### or do it by super().buy(), ## in py2 -> super(MP3Player, self).buy()
# super will again follow MRO rules and will call the 1st base inherited class.

## Adapters is very good design pattern
# ducktyping is cheaper then inheritance.
```
